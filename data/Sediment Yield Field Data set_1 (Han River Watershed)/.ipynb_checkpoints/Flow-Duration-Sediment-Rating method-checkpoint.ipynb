# Han River Watershed

Daily discharge from 2005/1/1 to 2014/12/31.
Length of sediment records varies with stations.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

Load all excel files in the folder

```python
def findExcelFiles():
    files = []
    for file in os.listdir(os.curdir):        
        if file.endswith(".xlsx"):
            if file.startswith("Sediment"):
                files.append(file)
    return files 
```

Using daily discharge from 2005/1/1 to 2014/12/31. Missing data is excluded.
Discharge is ranked from largest to smallest, exceedence probability can then be calculated as follows:

$P = 100 \times [M/(n+1)]$

    P = the probability that a given flow will be equaled or exceeded (% of time)
    M = the ranked position on the listing (dimensionless)
    n = the number of events for period of record (dimensionless)

The discharges of interval mid-point are interploated or extrapolated linearly.

```python
def dailyDischarge(file):    
    def extrapolate(dataframe):
        nanIndex = dataframe.Q[np.isnan(dataframe.Q)==True].index.tolist()
        try:
            if len(nanIndex)>0:
                for i in nanIndex:
                    indexlist = dataframe.index.tolist()
                    j = indexlist.index(i)
                    q1 = dataframe.iat[j+1,0]
                    q2 = dataframe.iat[j+2,0]
                    p1 = dataframe.iat[j+1,1]
                    p2 = dataframe.iat[j+2,1]
                    p0 = dataframe.iat[j,1]
                    q0 = q2 - (p2-p0)*(q2-q1)/(p2-p1)
                    dataframe.loc[i,'Q'] = q0
            return dataframe
        except:
            return dataframe
    
    df = pd.read_excel(file, sheetname='stage-discharge(daily)',skiprows=3,na_values=["'-","-"])
    df = df.drop(df.columns[[0,2,5,8,11,14,17,20,23,26,29]], axis=1)
       
    df1 = df[df.columns[0]].append(df[df.columns[2]]).append(df[df.columns[4]]).append(df[df.columns[6]]).append(df[df.columns[8]]).append(df[df.columns[10]]).append(df[df.columns[12]]).append(df[df.columns[14]]).append(df[df.columns[16]]).append(df[df.columns[18]]).reset_index(drop=True)
    df2 = df[df.columns[1]].append(df[df.columns[3]]).append(df[df.columns[5]]).append(df[df.columns[7]]).append(df[df.columns[9]]).append(df[df.columns[11]]).append(df[df.columns[13]]).append(df[df.columns[15]]).append(df[df.columns[17]]).append(df[df.columns[19]]).reset_index(drop=True)
    
    df = pd.concat([df1,df2], axis=1).dropna()
    df.columns = ['date', 'Q']
    
    df_sorted = df.sort_values('Q',ascending=False)
    df_sorted.loc[:,'rank'] = df_sorted['Q'].rank(ascending=False,method='min')
    df_sorted.loc[:,'P'] = df_sorted['rank']/(df_sorted.date.count()+1) * 100.0

    df_sorted = df_sorted.drop(df_sorted.columns[[0,2]], axis=1)
    
    mid = [0.03,0.08,0.3,1,3.25,10,20,30,40,50,60,70,80,90,97.5]
    d = [0.06,0.04,0.4,1,3.5,10,10,10,10,10,10,10,10,10,5]
    a = np.empty(len(mid))
    a[:] = np.nan
    d_ = {'Q':a,'P':mid}
    df2 = pd.DataFrame(data=d_)
       
    result = pd.concat([df_sorted,df2]).drop_duplicates('P')
    result = result.sort_values('P')
    result2 = result.set_index('P')
    result2 = result2.interpolate(method='index')
    result2.loc[:,'P'] = result2.index
       
    result2 = result2.reset_index(drop=True)

    extrapolate(result2)
    
    selected = result2.loc[result2['P'].isin(mid)]
    selected.loc[:,'D'] = d
    return result2, selected
```

The given total load-discharge relationships are given year by year. Here regression was done by using all available discharge and total load.

```python
def sedimentRating(file):
    df = pd.read_excel(file,sheetname='sediment measurement',skiprows=4,parse_cols=[11,23])
    df1 = df.apply(pd.to_numeric, errors='coerce')
    df1.columns = ['Q', 'Qt']
    df1 = df1.dropna().sort_values('Q',ascending=False)
    xdata = np.array(df1.Q)
    ydata = np.array(df1.Qt)
    b, a = np.polyfit(np.log(xdata),np.log(ydata),1)
    a = np.e**a
    return xdata, ydata, a, b
```

```python
files = findExcelFiles()
```

Flow Duration Curve

```python
plt.figure(figsize=(15,25))
for file in files:
    f = file[file.find('_')+1:file.rfind('_')]
    c = files.index(file)
    x,y,a,b = sedimentRating(file)
    res, sel = dailyDischarge(file)   
    plt.subplot(4,2,c+1)
    plt.plot(x,y,'o')
    plt.plot(x,a*x**b)
    plt.xscale('log')
    plt.yscale('log')
    a_string = '%s' % float('%.4g' % a)
    b_string = '%s' % float('%.4g' % b)
    func_string = r'${}x^{}$'.format(a_string,b_string)
    
    #plt.annotate(s=func_string, ha='center')
    plt.xlabel('Q (cms)')
    plt.ylabel(r'$Q_t$ (tons/day)')
    plt.title(f)
```

Sediment Rating Curve is plotted as below.

```python
plt.figure(figsize=(15,25))

for file in files:
    f = file[file.find('_')+1:file.rfind('_')]
    c = files.index(file)
    x,y,a,b = sedimentRating(file)
    res, sel = dailyDischarge(file) 
    plt.subplot(4,2,c+1)
    plt.plot(res.P, res.Q)
    plt.plot(sel.P, sel.Q, 'o')
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('% time exceeded')
    plt.ylabel('Dischage (cms)')
    plt.title(f)
    plt.grid(True,which='major')
```

```python
nameList = []
sedimentYield = []
area = [11074.05, 283.46, 1346.01, 173.43, 518.57, 8822.74, 306.66]

for file in files:
    f = file[file.find('_')+1:file.rfind('_')]
    c = files.index(file)
    x,y,a,b = sedimentRating(file)
    res, sel = dailyDischarge(file)   
    sel.loc[:,'Qt'] = a*sel.Q**b
    sel.loc[:,'QtxP'] = sel.Qt * sel.D/100
    nameList.append(f)
    sedimentYield.append(sel.QtxP.sum()*365)
    print(sel)
    print('The sediment yield of {} is {} tons/year'.format(f, round(sel.QtxP.sum()*365,2)))
```

Q is discharge in cubic meter per second; P is interval midpoint (%); D is interval (%); Qt is sediment total load in tons per day. QtxP is sediment total load times interval in tons per day.

```python
print('Summary table:')
YT = pd.DataFrame({'Name':nameList, 'Sediment':sedimentYield})
YT.loc[:,'Area'] = area
YT['Sediment'] = YT['Sediment']/YT.Area
print(YT)

```

The above table shows the sediment yield of each sub-watershed. "Name" is the code of the station. "Sediment" is sediment yield in tons/$km^2$/year. Area is in $km^2$.

```python
plt.figure(figsize=(10,6))
plt.plot(YT.Area, YT.Sediment, 'o')
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$Drainage Area (km^2)$',fontsize=14)
plt.ylabel(r'$Specific degradation (ton/km^2\cdot year)$',fontsize=14)
```
